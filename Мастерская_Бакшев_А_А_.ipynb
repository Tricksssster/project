```python
hard_skills_da = vacancies_da['hard_skills_from_description'].str.split(',').explode().str.strip()

top_hard_skills_by_experience = hard_skills_da.groupby(vacancies_da['experience']).value_counts().groupby(level=0).head(10)

top_hard_skills_by_experience.unstack().plot(figsize=(15,6), kind='bar', stacked=False)
plt.title('Топ хард скиллов по грейдам опыта работы DA')
plt.xlabel('Опыт работы')
plt.ylabel('Частота')
plt.legend(title='хард-скиллы')
plt.xticks(rotation=0)
plt.show()
```

```python
employment_da = vacancies_da['employment']

employment_da_by_experience = employment_da.groupby(vacancies_da['experience']).value_counts().groupby(level=0).head(5)

employment_da_by_experience.unstack().plot(figsize=(15,10), kind='bar', stacked=False)
plt.title('Предложения по занятости для DA')
plt.xlabel('Опыт работы')
plt.ylabel('Количество вакансий')
plt.xticks(rotation=0)
plt.legend(title='занятость')
plt.show()

employment_ds = vacancies_ds['employment']

employment_ds_by_experience = employment_ds.groupby(vacancies_ds['experience']).value_counts().groupby(level=0).head(5)

employment_ds_by_experience.unstack().plot(figsize=(15,10), kind='bar', stacked=False)
plt.title('Предложения по занятости для DS')
plt.xlabel('Опыт работы')
plt.ylabel('Количество вакансий')
plt.xticks(rotation=0)
plt.legend(title='занятость')
plt.show()
```

В части занятости вакансий для ДА можно сделать вывод, что абсолютное большинство компаний рассматривает полную занятость с редкими исключениями.

В разрезе ДС дела обстоят чуть интереснее, но по сути аналогично. Джунов чаще зовут на стажировки и проектные работы, чтобы включить в работу и ввести в курсы дел компаний. Это хороший и частый опыт.

```python
schedule_da = vacancies_da['schedule']

schedule_da_by_experience = schedule_da.groupby(vacancies_da['experience']).value_counts().groupby(level=0).head(5)

schedule_da_by_experience.unstack().plot(figsize=(15,10), kind='bar', stacked=False)
plt.title('Предложения по графику для DA')
plt.xlabel('Опыт работы')
plt.ylabel('Количество вакансий')
plt.legend(title='график')
plt.xticks(rotation=0)
plt.show()

schedule_ds = vacancies_ds['schedule']

schedule_ds_by_experience = schedule_ds.groupby(vacancies_ds['experience']).value_counts().groupby(level=0).head(5)

schedule_ds_by_experience.unstack().plot(figsize=(15,10), kind='bar', stacked=False)
plt.title('Предложения по графику для DS')
plt.xlabel('Опыт работы')
plt.ylabel('Количество вакансий')
plt.legend(title='график')
plt.xticks(rotation=0)
plt.show()
```

Опиряйсь на графики, мы можем увидеть следующее факт:

- DA
Джуниоров рассматривают в большинстве случаев на полный рабочий день, но иногда бывают исключения на сменный график. У джунов с опытом, мидлов и сеньёров чаще всего попадаются вакансии с полным рабочим днем, но есть неплохая доля предложений с возможностью удаленной работы.


- DS
У ДС примерно все также, как и у ДА. Преобладает полный рабочий день, иногда всречаются предложения с удаленным формат и гибким графиком.


Перейдем к рассмотрению периодичности публикации объявлений по вакансиям. У нас всего два месяца, так что лучше сделаем разбивку по дням, а не по месяцам

```python
vacancies_da['day'] = vacancies_da['published_at'].dt.to_period('D')

daily_vacancies_by_experience = vacancies_da.groupby(['day', 'experience']).size().unstack().fillna(0)

daily_vacancies_by_experience.plot(figsize=(12, 6), marker='h', markerfacecolor='g', alpha=1)
plt.title('Динамика размещений вакансий DA')
plt.xlabel('Дата')
plt.ylabel('Число вакансий')
plt.xticks(rotation=0)
plt.legend(title='опыт работы')
plt.grid(True)
plt.show()
```

Судя по графику, мы можем увидеть, что активность по размещению вакансий значительно увеличилась к концу апреля. Сеньёрово  и джунов без опыта ищут гораздо меньше, чем других специалистов, и их динамика, скорее всего ни от чего не зависит.

С мидлами и джунами дела обстоят интереснее. Наблюдается и падение, и резкий подъем в количестве вакансий. Если учесть, что начало мая это уже более или менее отпускной сезон, то можно сказать, что компании сталкиваются с проблемами отсутствия специалистов, уходящих в отпуск, и начинают оперативный поиск сотрудников на вакантные имеющиеся места. Это как раз и касается наиболее широких групп сотрудников - джунов и мидлов.

```python
vacancies_ds['day'] = vacancies_ds['published_at'].dt.to_period('D')

daily_vacancies_by_experience_ds = vacancies_ds.groupby(['day', 'experience']).size().unstack().fillna(0)

daily_vacancies_by_experience_ds.plot(figsize=(12, 6), marker='h', markerfacecolor='g', alpha=1)
plt.title('Динамика размещений вакансий DS')
plt.xlabel('Дата')
plt.ylabel('Число вакансий')
plt.xticks(rotation=0)
plt.legend(title='опыт работы')
plt.grid(True)
plt.show()
```

По сути ситуация такая же, как и с вакансиями ДА.
Резкий скачок под конц апреля. Сеньёров и джунов без опыта ищут меньше, чем остальных. Предложений по широкому кругу сотрудников - джунов с опытом и мидлов гораздо больше среди остальных. Пиковые значения наблюдаются под конец месяца.

Кстати, как еще один фактор скачка может быть начало нового отчетного месяца. Заработную плату расчитытывать проще и оформлять нового сотрудника, если он устаривается с 1 числа месяца. Возможно, поэтому поиск сотрудников и активизируется, чтобы успеть его оформить к началу нового трудового месяца)
